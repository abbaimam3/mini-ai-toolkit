```python
# Install libraries
!pip install spacy textblob matplotlib
!python -m spacy download en_core_web_sm

# Import libraries
import spacy
from textblob import TextBlob
import matplotlib.pyplot as plt

# Load NLP model
nlp = spacy.load("en_core_web_sm")

# Sample Amazon-style product reviews
reviews = [
    "I absolutely love the new Samsung Galaxy S23! The camera quality is outstanding.",
    "The battery of my Lenovo laptop dies too quickly, very disappointed.",
    "Apple AirPods are great but a bit overpriced for the sound quality.",
    "I bought a Sony headphone last week and it‚Äôs simply amazing!",
    "This HP printer is the worst purchase I‚Äôve ever made. Constant paper jams!",
    "The Dell monitor works perfectly, excellent brightness and design!",
    "Terrible experience with my Xiaomi phone, lags every time I open an app.",
    "The Bose speaker sound quality is crystal clear ‚Äî totally worth the money!",
    "My Asus laptop performance is average, nothing special.",
    "Amazon delivery was fast, but the Logitech mouse stopped working after a week."
]

# Sentiment & entity extraction
sentiment_summary = {"Positive": 0, "Negative": 0, "Neutral": 0}

for review in reviews:
    doc = nlp(review)
    print(f"\nüìù Review: {review}")

    # Extract named entities
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    print("üîç Named Entities:", entities if entities else "No named entities found.")

    # Sentiment Analysis
    sentiment = TextBlob(review).sentiment.polarity
    if sentiment > 0.05:
        sentiment_label = "Positive"
    elif sentiment < -0.05:
        sentiment_label = "Negative"
    else:
        sentiment_label = "Neutral"

    sentiment_summary[sentiment_label] += 1
    print(f"üí¨ Sentiment: {sentiment_label} (Score: {sentiment:.2f})")

# Visualization ‚Äî Sentiment Distribution
plt.figure(figsize=(6, 4))
plt.bar(sentiment_summary.keys(), sentiment_summary.values())
plt.title("Sentiment Distribution of Amazon Reviews")
plt.xlabel("Sentiment Category")
plt.ylabel("Number of Reviews")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
